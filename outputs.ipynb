Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a comprehensive data visualization library in Python. It provides a wide range of tools for constructing plots, charts, and visualizations, making it a popular choice for data scientists and analysts.
Matplotlib is used to create static, animated, and interactive visualizations in Python. It enables users to generate high-quality charts and graphs for better data understanding and presentation. It supports a variety of plot types and customization options, making it versatile for different data visualization needs.

Five plots that can be plotted using the Pyplot module of Matplotlib:

Line Plot:

Used to represent data points with connected lines, suitable for showing trends.

Bar Plot:

Displays rectangular bars of lengths proportional to the values they represent, suitable for comparing categories.

Scatter Plot:

Represents individual data points on a two-dimensional plane, helpful for exploring relationships between variables.

Histogram:

Visualizes the distribution of a dataset by dividing it into bins and displaying the frequency of values in each bin.

Pie Chart:

Represents data in a circular graph where each wedge corresponds to a proportion of the whole, useful for showing part-to-whole relationships.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.


```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
# Note: Also add title, xlabel, and ylabel to the plot.
```

```python
plt.plot(x,y)
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("This is title")
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

The subplot() function in Matplotlib is used to create multiple plots within a single figure. It allows you to arrange plots in a grid, specifying the number of rows and columns. This is beneficial when you want to compare different datasets or visualize multiple aspects of your data in a single figure.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for four line plots
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating subplots
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line Plot 4')

# Adjust layout for better spacing
plt.tight_layout()


plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot (or bar chart) is a type of chart that presents categorical data with rectangular bars. Each bar's length or height corresponds to the value it represents. Bar plots are used to visualize and compare the values of different categories, making them suitable for displaying the distribution or comparison of discrete data points.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar Plot
plt.figure(figsize=(5, 3))
plt.bar(company, profit, color=['#FF9C86', '#FFE986', '#DDE857', '#E8A057'])
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.title('Company Profits Bar Plot')
plt.show()

# Horizontal Bar Plot
plt.figure(figsize=(5, 3))
plt.barh(company, profit, color=['#5792E8', '#7A57E8', '#AD57E8', '#FFE986'])
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.title('Company Profits Horizontal Bar Plot')
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.


A box plot, also known as a box-and-whisker plot, is a statistical visualization tool used to display the distribution of a dataset. It shows the median, quartiles, and potential outliers in a concise manner. Box plots are useful for identifying the central tendency, spread, and skewness of data, making them valuable for comparing different datasets or examining the variability within a single dataset.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Creating a box plot
plt.figure(figsize=(8, 5))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot of Two Datasets')
plt.show()
```
